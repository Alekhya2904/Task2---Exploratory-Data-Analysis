```python
import pandas as pd
df=pd.read_csv("unzipped_folder/Titanic-Dataset.csv")
print(df.head())
print(df.info())
print(df.describe())
print(df.median(numeric_only=True))
print(df.isnull().sum())
import matplotlib.pyplot as plt
df['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x=df['Age'])
plt.title("Age Boxplot")
plt.show()
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
sns.pairplot(df[['Age','Fare','Survived','Pclass']])
plt.show()
```

## Age Distribution Analysis:
- Most passengers were between 20 and 40 years old.
- The highest concentration of passengers was observed around 20–30 years.
- The number of passengers decreases as age increases.
- Very few passengers were above 70 years of age.
- The distribution is slightly right-skewed, indicating a small number of older passengers.                    
## Age Boxplot Observations:
- The median age of passengers is around 28 years.
- Most passenger ages fall between 20 and 38 years.
- A few passengers above 65 years appear as outliers.
- The maximum age is close to 80 years.
- The distribution is slightly right-skewed due to older passengers.
## Correlation Heatmap Observations
- Survival has a positive correlation with Fare, indicating that passengers who paid higher fares had a better chance of survival.
- Survival has a negative correlation with Pclass, meaning first-class passengers were more likely to survive than third-class passengers.
- Fare and Pclass show a negative correlation because higher-class passengers generally paid higher fares.
- Age has a weak correlation with most other features, suggesting that age alone is not a strong predictor of survival.
- No extremely high correlations were observed, indicating low multicollinearity among the features.
## Pairplot Observations
- Passengers with higher fares tend to have a higher survival rate.
- First-class passengers show better survival outcomes compared to lower classes.
- Most passengers are concentrated in the age range of 20–40 years.
- The Fare feature contains a few extreme values (outliers).
- Age and Fare do not show a strong linear relationship.
- The pairplot helps visualize the distribution of features and relationships between them.
## Conclusion
Exploratory Data Analysis was performed on the Titanic dataset using statistical summaries and visualizations. Most passengers were between 20 and 40 years old. A few elderly passengers appeared as outliers in the Age feature. Higher-class passengers and those who paid higher fares had better survival rates. Correlation analysis showed that Fare and Pclass were related to survival, while Age had a weaker impact. Visualizations helped identify patterns, trends, and anomalies within the dataset.
